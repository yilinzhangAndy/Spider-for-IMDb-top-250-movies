# Overall result plot

### 1. Step 1: Import data
- Import data obtained through Spider (IMDb.xls)

```python
from spider_tools.Data_Visualization import Data
raw_data = Data('./data/IMDb.xls')
data = raw_data.Import()
```

### 2. Step 2: Data cleaning
- Clean the day, month, and text in the year
- Clean symbols in Gross
- Clean the name of Countries, If it is jointly filmed by multiple countries, the first country will be used as the representative

```python
clean_data = raw_data.Cleaning(data) 
clean_data
```

### 3. Step 3: Data Statistics
- Divide the year of the movie into chronology  
- Count the number of countries

```python
sta = raw_data.Statistics(clean_data)

#the list what we need for visualization after data statistics
```

### 4. Step 4:Data Visualization
- Use matplotlib to show the Distribution of Country origin
- Move the largest part away from the center of the circle
- Show the percentages of each part

```python
raw_data.Visualization(clean_data,sta )
```
